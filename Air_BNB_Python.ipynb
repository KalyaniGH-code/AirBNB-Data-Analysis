# Data Analysis

Univariate Analysis - Analyzing each of these single columns and their data distributions

```python
# Identifying Outliers
df01 = df[df['price']<1500]
sns.boxplot(data = df01, x = 'price')
```

```python
# Price Distribution Analysis
plt.figure(figsize=(10,5))
sns.histplot(data = df01, x = 'price', bins= 100)
plt.ylabel('Frequency')
plt.xlabel('Price')
plt.title('Price Distribution Analysis')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data = df01, x = 'availability_365', bins=50, color='orange')
plt.ylabel('Frequency')
plt.title('availability_365 Analysis')
plt.show()
```

Bivariate Analysis

```python
df01.groupby('neighbourhood_group')['price'].mean()
```

# Feature Engineering
Creating a New Column

```python
# df01['Price per Bed']
df01['price per bed'] = df01['price']/df01['beds']
df01.head()
```

```python
# Average Price Per Bed
df01.groupby('neighbourhood_group')['price per bed'].mean()
```

Bivariate Analysis- One variable depedency on another column

```python
df01.columns
```

```python
# Price Depedencies on Neighbourhood
plt.figure(figsize=(8, 6))
sns.barplot(data= df01, x= 'neighbourhood_group', y='price', saturation=0.75, palette='viridis', hue='room_type')
plt.title('Bivariate Analysis for Neighbourhood_group Vs Price')
plt.xlabel('Neighbourhood_group')
plt.ylabel('Price')
plt.show()
```

```python
# No of Reviews with Price Relationship
plt.figure(figsize=(5,5))
sns.scatterplot(data= df01, x='number_of_reviews', y='price', hue='neighbourhood_group', palette='deep')
plt.title('Locality and Review Depedency')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

```python
df01.columns
```

```python
sns.pairplot(data=df01, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'],hue='room_type', palette='Spectral')
```

```python
# Geographical Distribution of AirBNB Listing as we have Latitude and Longitude available
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data= df01, x='longitude', y='latitude', hue='room_type')
plt.title('Geographical Distribution of AirBNB Listings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Heatmap- Correlation of one variable with other
order = df01[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
       'longitude', 'latitude', 'availability_365','beds']].corr()
order
```

```python
sns.heatmap(data= order, annot=True, cmap='RdPu',linewidths=0.5, linecolor="black")
plt.show()
```
